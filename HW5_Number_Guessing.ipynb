```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
def binary_search(N, x):
    """ Perform a binary search in the range [1, N] to find the number x.
    Returns the number of steps taken to find x. """
    low = 1
    high = N
    steps = 0

    while low <= high:
        mid = (low + high) // 2
        steps += 1
        if mid == x:
            return steps
        elif mid < x:
            low = mid + 1
        else:
            high = mid - 1

    return steps
```

```python
def simulate_binary_search(max_N, simulations_per_N):
    """ Simulate the binary search for various values of N and average the number of questions. """
    avg_steps = []

    for N in range(1, max_N + 1):
        total_steps = 0
        for _ in range(simulations_per_N):
            x = np.random.randint(1, N + 1)  # Random target in the range [1, N]
            total_steps += binary_search(N, x)
        avg_steps.append(total_steps / simulations_per_N)

    return avg_steps
```

```python
# Parameters for simulation
max_N = 1000  # Maximum value of N to simulate
simulations_per_N = 100  # Number of simulations per value of N
```

```python
# Run the simulation
avg_steps = simulate_binary_search(max_N, simulations_per_N)
```

```python
# Plotting the results
plt.figure(figsize=(12, 6))
plt.plot(range(1, max_N + 1), avg_steps, color='blue', marker='o')
plt.title("Average Number of Questions in Binary Search vs. N")
plt.xlabel("N (Size of search space)")
plt.ylabel("Average Number of Questions")
plt.grid(True)
plt.show()
```
