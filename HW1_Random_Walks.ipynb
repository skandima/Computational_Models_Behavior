```python
import matplotlib.pyplot as plt
import random 
import numpy as np
```

```python


def random_walk(T):
    x = 0 #initialize state of random walks
    t = 0 #initialize counter for number of steps
    positions = [x]

    while t < T:
        y = random.choice([0,1]) #simulate a sample from a fair coin
        if y == 0:
            x += 1
        else:
            x -= 1
        
        positions.append(x)
        t += 1

    return positions

T = 100
num_walks = 5

plt.figure(figsize= (10,6))
for i in range(num_walks):
    positions = random_walk(T)
    plt.plot(positions, label = f'Walk {i+1}')

plt.title('Random Walks')
plt.xlabel('Steps')
plt.ylabel('Position')
plt.legend()
plt.grid(True)
plt.show

```

```python
def gaussian_random_walk(T):
    x = 0 #initialize state of random walks
    t = 0 #initialize counter for number of steps
    positions = [x]

    for _ in range(T):
        step = np.random.randn()  # Get a sample from a standard normal distribution
        x += step
        positions.append(x)

    return positions

T = 100
num_walks = 5

plt.figure(figsize= (10,6))
for i in range(num_walks):
    positions = gaussian_random_walk(T)
    plt.plot(positions, label = f'Walk {i+1}')

plt.title('Random Walks - Gaussian')
plt.xlabel('Steps')
plt.ylabel('Position')
plt.legend()
plt.grid(True)
plt.show
```
